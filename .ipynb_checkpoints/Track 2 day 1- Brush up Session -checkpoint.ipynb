#armstrong number

It's a number that equals the sum of its digits, each raised to a power. For example, if you have a number like 153, it's an Armstrong number because 1^3 + 5^3 + 3^3 equals 153.

Armstrong Numbers:
153, 370, 371, 407, 1634, 8208

```python
def find_armstrong(start,end):
    armstr=[]
    for num in range(start,end+1):
        if is_armstrong(num):
            armstr.append(num)
    return armstr
find_armstrong(3,44)
```

```python
def is_armstrong_number(number):
    num_digits = len(str(number))
    total = sum(int(digit) ** num_digits for digit in str(number))
    return total == number
 
def find_armstrong_numbers(start, end):
    armstrong_numbers = []
    for num in range(start, end + 1):
        if is_armstrong_number(num):
            armstrong_numbers.append(num)
    return armstrong_numbers
```

```python
n = input("number- ")
len_n = len(n)
arm = sum(int(i) ** len_n for i in n)
if int(n) == arm:
    print(n, "is an armstrong number")
else:
    print(n, "not an armstrong number")
```
