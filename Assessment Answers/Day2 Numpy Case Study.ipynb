```python
users=100
movies=1000
movie_matrix=np.random.randint(-1,11,size=(users,movies))
movie_matrix
```

```python
df=pd.DataFrame(movie_matrix,columns=movie_id)
df
```

```python
# Appended users=10 movies=50
new_movie_id=np.arange(2301,2351)
```

```python
new_movie_user=np.arange(101,111)
#new_movie_user
df
```

```python
new_movie_matrix=np.random.randint(0,11,size=(10,50))
new_movie_matrix
```

```python
for i in range(50):
    df[new_movie_id[i]] = np.random.randint(-1,11,size=(100,1050))  # Merging new movie id with rating =-1 in old dataframe
```

```python
new_movie_matrix.shape
```

```python
df
```

```python
for i in range(10):
    expert_ratings = new_movie_matrix[i]
    random_movie_indexes = np.random.choice(range(50), size=10, replace=False)

    for j in range(10):
        df.at[random_movie_indexes[j], new_movie_id[i]] = expert_ratings[j]
        
df
```

```python
#final movie matrix with four col.
final_movie_rating=pd.DataFrame(df,columns=['Movie ID', 'Average rating', 'Number of ratings', 'Standard deviation of ratings'])
final_movie_rating
```

```python
df.columns
```

```python
final_movie_rating['Movie ID']=df.columns
```

```python
for movie_id in df.columns:
    if movie_id not in final_movie_rating['Movie ID'].values:
        ratings = random.randint(0,1000)
        avg_rating = np.mean(ratings)
        num_ratings = random.randint(0,1000)
        std_dev = np.std(ratings)
        final_movie_rating = final_movie_rating.append({'Movie ID': movie_id,
                                                            'Average rating': avg_rating,
                                                            'Number of ratings': num_ratings,
                                                            'Standard deviation of ratings': std_dev},
                                                           ignore_index=True)

# Display the final_movie_rating matrix
final_movie_rating.head(10)
```
