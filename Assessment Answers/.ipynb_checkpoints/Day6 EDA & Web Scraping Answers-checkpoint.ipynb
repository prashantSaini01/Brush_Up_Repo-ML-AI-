# ******Univariate EDA******

```python

import pandas as pd
import seaborn as sns
cereal = pd.read_csv("cereal.csv")
cereal
```

```python
#Info function
cereal.info()  # We get 3 Categorical columns, 8 (INT) Numerical columns, 5 (FLOAT) Numerical columns
```

```python
#Describe function
cereal.describe() # Normal description about the data like Count,Mean,Std,Min,Percentiles,Max
```

```python
#Replace function
cereal["mfr"] = cereal["mfr"].replace({"N":"Nabisco",'A':'American Home Food Products','G':'General Mills','K':'Kelloggs', 'P':'Post', 'Q':'Quaker Oats', 'R':'Ralston Purina'})
cereal
```

```python
#Sorting Manufacturer based on Rating Top 5
sort = cereal.groupby(['mfr'])['rating'].mean()
sort.sort_values(ascending=False).head(5)
```

```python
# Barplots based on Manufacturer and Rating
import matplotlib.pyplot as plt
plt.figure(figsize = (12,6))
x = cereal['mfr']
y = cereal['rating']
plt.bar(x,y,color='purple')
plt.show()
```

```python
cereal
```

```python
# Pie Chart
pie = cereal["mfr"].value_counts()
pie
```

```python
col = cereal['mfr'].unique()
col
```

```python
fig = plt.figure(figsize=(12,6))
plt.pie(pie, autopct="%.2f",labels=col,explode=(0.1, 0.0, 0.2, 0.3, 0.0, 0.0,0.5),shadow=True)
plt.show()
```

```python
#Scatterplot between protein and calories
plt.scatter(cereal["protein"], cereal["calories"], cmap='viridis',c =[i for i in range(0,77)])
plt.colorbar()
plt.show()
```

```python
#Slicing nutrients for heatmap
nutrients = cereal.iloc[:,3:12]
```

```python
#Correlation
corr = nutrients.corr()
corr
```

```python
#Heatmap for correlation
figure = plt.figure(figsize=(12,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#Covariance
cov = nutrients.cov()
cov
```

```python
#Heatmap for Covariance
figure = plt.figure(figsize=(12,10))
sns.heatmap(cov,annot=True)
plt.show()
```

# ******Web Scraping******

```python
import requests
from bs4 import BeautifulSoup 
!pip install lxml
web = requests.get("https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population")
web
```

```python
web.content
```

```python
soup = BeautifulSoup(web.content,'lxml')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
c = soup.find(id="mw-content-text")
```

```python
content = c.find('table',class_="wikitable sortable sticky-header col2left col6left")
content
```

```python
data = content.find_all('tr')
r = []
for row in data[1:]:
    col = row.find_all(['th','td'])
    col = [i.text.split() for i in col]
    r.append(col)

df = pd.DataFrame(r,columns =['0','Country/Dependency','Population','% of world','Date','Source','6'])
df.drop(['0','6'],axis=1,inplace=True)
df
```
