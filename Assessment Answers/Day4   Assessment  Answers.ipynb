```python
#Day 4 Assessment 
#Subplot
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
iris = load_iris()
df  = pd.DataFrame(data=iris.data,columns = iris.feature_names )
```

```python
df
```

```python
# We use this to plot many different plots on the basis of given features
x = df['sepal length (cm)']
y = df['petal length (cm)']
a = np.array(x)
b=np.array(y)
w = df['sepal width (cm)']
z = df['petal width (cm)']
c = np.array(w)
d = np.array(z)
plt.subplot(1,2,1) # This decides the position of the following graph (Here it is 1 row,2 column,1st position)
plt.plot(a,b)
plt.subplot(1,2,2) # This decides the position of the following graph (Here it is 1 row,2 column,2nd position)
plt.plot(c,d)
plt.show()
```

```python
# For plotting subplots on top of one another we use (2,1,1) & (2,1,2)
plt.subplot(2,1,1) # This decides the position of the following graph (Here it is 2 row,1 column,1st position)
plt.plot(a,b)
plt.subplot(2,1,2) # This decides the position of the following graph (Here it is 2 row,1 column,2nd position)
plt.plot(c,d)
plt.show()
```

```python
# Grid Spec used to make grid plots using the help of various subplots in them
import matplotlib.gridspec as gridspec
plt.figure(figsize=(8,5))
gs=gridspec.GridSpec(2,2)
#subplot1
plt.subplot(2,2,1)
plt.scatter(x=df['sepal length (cm)'],y=df['petal width (cm)'])
plt.xlabel('sepal length (cm)')
plt.ylabel('petal width (cm)')
#subplot2
plt.subplot(2,2,2)
plt.scatter(x=df['petal length (cm)'],y=df['sepal width (cm)'])
plt.xlabel('petal length (cm)')
plt.ylabel('sepal width (cm)')
#subplot3
plt.subplot(2,2,3)
plt.scatter(x=df['sepal length (cm)'],y=df['sepal width (cm)'])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
#subplot4
plt.subplot(2,2,4)
plt.scatter(x=df['petal length (cm)'], y=df['petal width (cm)'])
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')

plt.tight_layout()

plt.show()
```
