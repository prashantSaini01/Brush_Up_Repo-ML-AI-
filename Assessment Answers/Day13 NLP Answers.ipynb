```python
import pandas as pd

# Read the 'mood_data.txt' file into a DataFrame
df = pd.read_csv("mood_data.txt", delimiter=';',header=None)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df.columns=["sentences","emotion"]
```

```python
df
```

```python
stops=stopwords.words("english")
```

```python
stops
```

```python
lm=df["sentences"]
```

```python
for i in range(len(lm)):
  lm[i]=[t for t in lm[i].split() if t.lower() not in stops]
  lm[i]=" ".join(lm[i])
```

```python
df
```

```python
df["cleaned_text"]=lm
```

```python
import re
```

```python
features=df["cleaned_text"]
processed_feat=[]
for i in range(0,len(features)):
 processed= re.sub(r'\W',' ',str(features[i]))
 processed= re.sub(r'\s+[a-zA-Z]\s',' ',processed)
 processed=re.sub(r'^[a-zA-Z]\s+',' ',processed)
 processed=re.sub('\s+',' ',processed)
 processed_feat.append(processed.lower())
```

```python
df["processed"]=processed_feat
```

```python
df
```

```python
x=df['processed']
y=df['emotion']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=41)
```

```python
tfidf=TfidfVectorizer()
x_train_tf=tfidf.fit_transform(x_train)
x_test_tf=tfidf.transform(x_test)
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train_tf,y_train)
```

```python
y_pred=rfc.predict(x_test_tf)
```

```python
accuracy_score(y_test,y_pred)*100
```
