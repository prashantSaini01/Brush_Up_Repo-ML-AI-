```python
from nltk.corpus import stopwords
```

```python
stops=stopwords.words('english')
```

```python
def clean_text(token):
    #tokenize
    token=nltk.word_tokenize(token)
    token=token[4:]
    token=" ".join(token)
    #http remove
    token=re.sub("https","",token)
    token = [i for i in token if i not in string.punctuation]
    token="".join(token)
    #stopwords removal
    token = [t for t in token.split() if t.lower() not in stops]
    return " ".join(token)
```

```python
df['cleaned_tweets']=df['text'].apply(clean_text)
```

```python
df['cleaned_tweets'][1]
```

```python
features = df["cleaned_tweets"]
processed_feat=[]
for i in range(0,len(features)):
    # Special character
    processed=re.sub(r'\W'," ",str(features[i]))
    
    # Remove character from the beginning
    processed=re.sub(r'\s+[a-zA-Z]\s',' ',processed)
    
    # Single character at the start
    processed=re.sub(r'^[a-zA-Z]\s+',' ',processed)
    
    # Space 
    processed=re.sub(r'\s+',' ',processed)
    
    # Lower
    processed=processed.lower()
    
    # Append
    processed_feat.append(processed)
```

```python
processed_feat[0]
```

```python
df['processed_text']=processed_feat
```

```python
def generate_polarity(text):
    sent=TextBlob(text).sentiment
    return sent
```

```python
df['tweet_sentiment']=df["processed_text"].apply(generate_polarity)
```

```python
df['polarity']=df['tweet_sentiment'].apply(lambda x:x[0])
df['subjectivity']=df['tweet_sentiment'].apply(lambda x:x[1])
df
```

```python
df_encoded=[]
for i in df.polarity:
    if i == 0:
        i="neutral"
        df_encoded.append(i)
    elif i > 0:
        i="positive"
        df_encoded.append(i)
    else:
        i="neagtive"
        df_encoded.append(i)
```

```python
df['encoded_polarity'] = df_encoded
```
