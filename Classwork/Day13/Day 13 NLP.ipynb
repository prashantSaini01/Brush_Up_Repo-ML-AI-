```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel("Online_Retail.xlsx")
df
```

```python
df["Description"].unique()
```

```python
df["Country"].unique()
```

```python
df.isnull().sum()
```

```python
#Dropping Customer ID
df.drop("CustomerID",axis=1,inplace=True)
```

```python
# Removing white spaces from Description
df["Description"] = df["Description"].str.strip()
df
```

```python
df["InvoiceNo"].str.contains('C').sum()
```

```python
df["InvoiceNo"]=df["InvoiceNo"].astype("str")
```

```python
df = df[~df["InvoiceNo"].str.contains("C")]
df
```

```python
france=df[df["Country"]=="France"]
```

```python
france
```

```python
france.shape
```

```python
france.isnull().sum()
```

```python
basket = france.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
```

```python
basket
```

```python
def encode_data(x):
    if x<=0:
        return 0
    else:
        return 1
encoded_basket = basket.applymap(encode_data)
encoded_basket
```

```python
frequent_itemsets = apriori(encoded_basket,min_support=0.07,use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rules
```

```python
rules1 = association_rules(frequent_itemsets,metric="confidence",min_threshold=1)
```

```python
rules1
```

# Natural Language Processing (NLP)

```python
import nltk
nltk.download('punkt')
sent = "PCA allows for dimensionality reduction by retaining only a subset of the principal components that capture the majority of"
```

```python
# Tokenization
nltk.sent_tokenize(sent)
```

```python
tokens = nltk.word_tokenize(sent)
tokens
```

```python
from nltk.util import bigrams,trigrams
```

```python
list(bigrams(tokens))
```

```python
list(trigrams(tokens))
```

```python
# POS(Part Of Speech) tagging
nltk.pos_tag(tokens)
```

```python
sent= "Jim eats a banana"
```

```python
token = nltk.word_tokenize(sent)
token
```

```python
nltk.pos_tag(token)
```

```python
for i in token:
    print(nltk.pos_tag([i])) # This shows that eats is a Noun
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
reg = RegexpTokenizer('(?u)\W+|\$[\d\.]+|\S+')
r_tokens = reg.tokenize(sent)
for i in r_tokens:
    print(nltk.pos_tag([i])) 
```
