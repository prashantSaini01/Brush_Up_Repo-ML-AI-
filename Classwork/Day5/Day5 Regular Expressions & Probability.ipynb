```python
pattern = re.compile(r'abcd')
for i in pattern.finditer(text_to_search):  #matching is case sensitive
        print(i)
```

```python
pattern = re.compile(r'.') # it matches with all characters in our string
for i in pattern.finditer(text_to_search):  
        print(i)
```

```python
pattern = re.compile(r'\D')
for i in pattern.finditer(text_to_search): 
        print(i)
```

```python
pattern = re.compile(r'\s')
for i in pattern.finditer(text_to_search):  
        print(i)
```

```python
pattern = re.compile(r'\S')
for i in pattern.finditer(text_to_search):  
        print(i)
```

```python
pattern = re.compile(r'\bHa') # before comes space or a new line
for i in pattern.finditer(text_to_search):  #matching is case sensitive
        print(i)
```

```python
pattern = re.compile(r'\BHa') # matches with no space or new line
for i in pattern.finditer(text_to_search):  #matching is case sensitive
        print(i)
```

```python
sent ="My name is Adarsh Singh"
pattern = re.compile(r'^M')
for i in pattern.finditer(sent):
    print(i)
```

```python
pattern = re.compile(r'\d\d\d\D\d\d\d\D\d\d\d') #or (r'\d{3}.\d{3}.\d{3})
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
pattern = re.compile(r'\d\d\d[.-]\d\d\d[.-]\d\d\d') 
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
pattern = re.compile(r'[89]00[.-]\d\d\d[.-]\d\d\d') 
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
#to check for inndividual character we use (r'[^a-zA-Z]')
pattern = re.compile(r'[^a-zA-Z]') 
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
pattern = re.compile(r'[cmp]at') 
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
pattern = re.compile(r'Mr.') # this is an error for which we have to use quantifiers
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
#Quantifiers :-  +(1 or more),?( 0 or 1),*(0 or more)
pattern = re.compile(r'Mr\.?')
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
pattern = re.compile(r'Mr\.?\s[A-Z]')
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
pattern = re.compile(r'Mr\.?\s[A-Z]\w+')
for i in pattern.finditer(text_to_search):
    print(i)
```

```python
pattern = re.compile(r'(Mr|Ms|Mrs)\.?\s[A-Z]\w*')
for i in pattern.finditer(text_to_search):
    print(i)
```
